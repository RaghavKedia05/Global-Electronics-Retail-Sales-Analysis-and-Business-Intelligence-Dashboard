# Phase 1: Data Understanding

This notebook explores the six source CSV files. It displays the findings inside the notebook and does not create a separate report.

## 1. Import the required libraries

```python
from pathlib import Path
import pandas as pd
```

## 2. Set the data folder path

The code works when the notebook is opened from either the project folder or the `notebooks` folder.

```python
current_folder = Path.cwd()

if current_folder.name == 'notebooks':
    project_folder = current_folder.parent
else:
    project_folder = current_folder

data_folder = project_folder / 'data' / 'raw'
print('Data folder:', data_folder)
```

## 3. Load the CSV files

`Customers.csv` uses Windows-1252 encoding. The remaining files use UTF-8.

```python
customers = pd.read_csv(data_folder / 'Customers.csv', encoding='cp1252', keep_default_na=False, na_values=[''])
products = pd.read_csv(data_folder / 'Products.csv', keep_default_na=False, na_values=[''])
sales = pd.read_csv(data_folder / 'Sales.csv', keep_default_na=False, na_values=[''])
stores = pd.read_csv(data_folder / 'Stores.csv', keep_default_na=False, na_values=[''])
exchange_rates = pd.read_csv(data_folder / 'Exchange_Rates.csv', keep_default_na=False, na_values=[''])
data_dictionary = pd.read_csv(data_folder / 'Data_Dictionary.csv', keep_default_na=False, na_values=[''])

print('All CSV files loaded successfully.')
```

## 4. Preview the datasets

```python
display(sales.head())
display(customers.head())
display(products.head())
display(stores.head())
display(exchange_rates.head())
```

## 5. Count rows and columns

A dictionary keeps each table name together with its DataFrame. This makes it easy to repeat the same check for every dataset.

```python
datasets = {
    'Customers': customers,
    'Products': products,
    'Sales': sales,
    'Stores': stores,
    'Exchange Rates': exchange_rates,
    'Data Dictionary': data_dictionary,
}

dataset_sizes = []

for dataset_name, dataframe in datasets.items():
    dataset_sizes.append({
        'Dataset': dataset_name,
        'Rows': dataframe.shape[0],
        'Columns': dataframe.shape[1],
    })

dataset_sizes = pd.DataFrame(dataset_sizes)
display(dataset_sizes)
```

## 6. Check column names and data types

```python
for dataset_name, dataframe in datasets.items():
    print(f'\n{dataset_name}')
    print(dataframe.dtypes)
```

## 7. Check missing values

```python
missing_value_results = []

for dataset_name, dataframe in datasets.items():
    for column_name in dataframe.columns:
        missing_count = dataframe[column_name].isna().sum()
        missing_percent = (missing_count / len(dataframe)) * 100

        if missing_count > 0:
            missing_value_results.append({
                'Dataset': dataset_name,
                'Column': column_name,
                'Missing Values': missing_count,
                'Missing Percentage': round(missing_percent, 2),
            })

missing_values = pd.DataFrame(missing_value_results)
display(missing_values)
```

## 8. Check complete duplicate rows

```python
for dataset_name, dataframe in datasets.items():
    duplicate_count = dataframe.duplicated().sum()
    print(f'{dataset_name}: {duplicate_count} duplicate rows')
```

## 9. Check primary keys

Sales uses a combined key because one order can contain several line items.

```python
customer_key_duplicates = customers.duplicated(subset=['CustomerKey']).sum()
product_key_duplicates = products.duplicated(subset=['ProductKey']).sum()
store_key_duplicates = stores.duplicated(subset=['StoreKey']).sum()
sales_key_duplicates = sales.duplicated(subset=['Order Number', 'Line Item']).sum()
rate_key_duplicates = exchange_rates.duplicated(subset=['Date', 'Currency']).sum()

print('Duplicate customer keys:', customer_key_duplicates)
print('Duplicate product keys:', product_key_duplicates)
print('Duplicate store keys:', store_key_duplicates)
print('Duplicate sales keys:', sales_key_duplicates)
print('Duplicate exchange-rate keys:', rate_key_duplicates)
```

## 10. Convert date columns

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'], errors='coerce')
sales['Delivery Date'] = pd.to_datetime(sales['Delivery Date'], errors='coerce')
customers['Birthday'] = pd.to_datetime(customers['Birthday'], errors='coerce')
stores['Open Date'] = pd.to_datetime(stores['Open Date'], errors='coerce')
exchange_rates['Date'] = pd.to_datetime(exchange_rates['Date'], errors='coerce')

print('Date columns converted.')
```

## 11. Find the sales date range and transaction counts

```python
first_order_date = sales['Order Date'].min()
last_order_date = sales['Order Date'].max()
sales_line_count = len(sales)
order_count = sales['Order Number'].nunique()

print('First order date:', first_order_date.date())
print('Last order date:', last_order_date.date())
print('Sales lines:', sales_line_count)
print('Distinct orders:', order_count)
```

## 12. Check relationships between tables

```python
missing_customer_keys = (~sales['CustomerKey'].isin(customers['CustomerKey'])).sum()
missing_product_keys = (~sales['ProductKey'].isin(products['ProductKey'])).sum()

physical_store_sales = sales[sales['StoreKey'] != 0]
missing_store_keys = (~physical_store_sales['StoreKey'].isin(stores['StoreKey'])).sum()
online_sales_lines = (sales['StoreKey'] == 0).sum()

print('Sales with missing customer:', missing_customer_keys)
print('Sales with missing product:', missing_product_keys)
print('Sales with missing physical store:', missing_store_keys)
print('Sales using StoreKey 0:', online_sales_lines)
```

## 13. Check exchange-rate coverage

```python
sales_with_rates = sales.merge(
    exchange_rates,
    how='left',
    left_on=['Order Date', 'Currency Code'],
    right_on=['Date', 'Currency'],
)

missing_exchange_rates = sales_with_rates['Exchange'].isna().sum()
print('Sales without an exchange rate:', missing_exchange_rates)
```

## 14. Check important sales rules

```python
missing_delivery_dates = sales['Delivery Date'].isna().sum()
delivery_before_order = (sales['Delivery Date'] < sales['Order Date']).sum()
non_positive_quantities = (sales['Quantity'] <= 0).sum()

print('Missing delivery dates:', missing_delivery_dates)
print('Delivery dates before order dates:', delivery_before_order)
print('Zero or negative quantities:', non_positive_quantities)
```

## 15. Create the dataset profiling summary

This summary answers the requirements for total records, total columns, and duplicate records.

```python
profiling_summary = []

for dataset_name, dataframe in datasets.items():
    profiling_summary.append({
        'Dataset': dataset_name,
        'Total Records': dataframe.shape[0],
        'Total Columns': dataframe.shape[1],
        'Duplicate Records': dataframe.duplicated().sum(),
    })

profiling_summary = pd.DataFrame(profiling_summary)
display(profiling_summary)
```

## 16. Create the column profiling table

This table shows the data type, missing values, missing percentage, and unique values for every field.

```python
column_profile_rows = []

for dataset_name, dataframe in datasets.items():
    for column_name in dataframe.columns:
        missing_count = dataframe[column_name].isna().sum()
        missing_percent = (missing_count / len(dataframe)) * 100

        column_profile_rows.append({
            'Dataset': dataset_name,
            'Column': column_name,
            'Data Type': str(dataframe[column_name].dtype),
            'Missing Values': missing_count,
            'Missing Percentage': round(missing_percent, 2),
            'Unique Values': dataframe[column_name].nunique(),
        })

column_profile = pd.DataFrame(column_profile_rows)
display(column_profile)
```

## 17. Display the supplied data dictionary

The data dictionary explains the meaning of each source field.

```python
display(data_dictionary)
```

## 18. Answer the Phase 1 questions

A sales row is one order line. The output therefore shows both order lines and distinct orders.

```python
print('PHASE 1 ANSWERS')
print('-' * 50)
print(f'Total sales records / order lines: {len(sales):,}')
print(f'Total distinct transactions / orders: {sales["Order Number"].nunique():,}')
print(f'Date range: {sales["Order Date"].min().date()} to {sales["Order Date"].max().date()}')
print()
print('Fields containing missing values:')

fields_with_missing_values = column_profile[column_profile['Missing Values'] > 0]

if fields_with_missing_values.empty:
    print('No missing values were found.')
else:
    display(fields_with_missing_values[[
        'Dataset',
        'Column',
        'Missing Values',
        'Missing Percentage',
    ]])
```
