# Phase 1: Data Understanding

## 1. Import libraries and set the data folder

```python
from pathlib import Path
import pandas as pd

project_folder = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data_folder = project_folder / 'data' / 'raw'
```

## 2. Load the datasets

```python
customers = pd.read_csv(data_folder / 'Customers.csv', encoding='cp1252', keep_default_na=False, na_values=[''])
products = pd.read_csv(data_folder / 'Products.csv', keep_default_na=False, na_values=[''])
sales = pd.read_csv(data_folder / 'Sales.csv', keep_default_na=False, na_values=[''])
stores = pd.read_csv(data_folder / 'Stores.csv', keep_default_na=False, na_values=[''])
exchange_rates = pd.read_csv(data_folder / 'Exchange_Rates.csv', keep_default_na=False, na_values=[''])
data_dictionary = pd.read_csv(data_folder / 'Data_Dictionary.csv', keep_default_na=False, na_values=[''])

datasets = {
    'Customers': customers,
    'Products': products,
    'Sales': sales,
    'Stores': stores,
    'Exchange Rates': exchange_rates,
}
```

## 3. Preview the datasets

```python
for dataset_name, dataframe in datasets.items():
    print(dataset_name)
    display(dataframe.head())
```

## 4. Check total records, columns, and duplicates

```python
profile_summary = []

for dataset_name, dataframe in datasets.items():
    profile_summary.append({
        'Dataset': dataset_name,
        'Total Records': len(dataframe),
        'Total Columns': len(dataframe.columns),
        'Duplicate Records': dataframe.duplicated().sum(),
    })

display(pd.DataFrame(profile_summary))
```

## 5. Check data types and missing values

```python
column_profile = []

for dataset_name, dataframe in datasets.items():
    for column_name in dataframe.columns:
        missing_count = dataframe[column_name].isna().sum()
        column_profile.append({
            'Dataset': dataset_name,
            'Column': column_name,
            'Data Type': str(dataframe[column_name].dtype),
            'Missing Values': missing_count,
            'Missing Percentage': round(missing_count / len(dataframe) * 100, 2),
        })

column_profile = pd.DataFrame(column_profile)
display(column_profile)
```

## 6. Answer the Phase 1 questions

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'])
fields_with_missing_values = column_profile[column_profile['Missing Values'] > 0]

print(f'Total sales records: {len(sales):,}')
print(f'Total transactions: {sales["Order Number"].nunique():,}')
print(f'Date range: {sales["Order Date"].min().date()} to {sales["Order Date"].max().date()}')
display(fields_with_missing_values)
```
